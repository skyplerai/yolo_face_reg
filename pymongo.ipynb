```python
from pymongo import MongoClient

# Connect to the MongoDB server
client = MongoClient('mongodb://localhost:27017/')  # Update the URI as needed

# Function to list all databases
def list_databases(client):
    databases = client.list_database_names()
    print("Available databases:")
    for db in databases:
        print(db)

# Usage
list_databases(client)
```

```python
db_name = 'face_recognition_db'
db = client[db_name]
```

```python
collections = db.list_collection_names()
print(collections)
```

```python
def view_db_contents(db):
    print("Collections in the database:")
    collections = db.list_collection_names()
    for collection_name in collections:
        print(f"\nCollection: {collection_name}")
        collection = db[collection_name]
        documents = collection.find()
        count = 0
        for doc in documents:
            print(doc)
            count += 1
        if count == 0:
            print(f"No documents found in the {collection_name} collection.")
```

```python
view_db_contents(db)
```

```python
def show_collection_sizes(db):
    collections = db.list_collection_names()
    print("Collection sizes:")
    for collection_name in collections:
        collection = db[collection_name]
        document_count = collection.estimated_document_count()  # Estimated count of documents
        print(f"Collection: {collection_name}, Size: {document_count} documents")
```

```python
show_collection_sizes(db)
```

```python
def delete_collection(db, collection_name):
    if collection_name in db.list_collection_names():
        db[collection_name].drop()
        print(f"Collection '{collection_name}' deleted.")
    else:
        print(f"Collection '{collection_name}' does not exist.")
```

```python
collection_name_to_delete = 'perm_faces'  # Replace with the collection you want to delete
delete_collection(db, collection_name_to_delete)
```

```python
client = MongoClient("mongodb://localhost:27017/")

# Select the database
db = client["test"]

# Select the collection
collection = db["new_collection"]

# Check the connection
print("Connection established to MongoDB:", client)
```

```python
# Insert a single document
single_doc = {"name": "Alice", "age": 25}
insert_result = collection.insert_one(single_doc)
print("Inserted document ID:", insert_result.inserted_id)
```

```python
# Insert multiple documents
multiple_docs = [
    {"name": "Bob", "age": 30},
    {"name": "Charlie", "age": 35},
    {"name": "David", "age": 40}]
insert_many_result = collection.insert_many(multiple_docs)
print("Inserted document IDs:", insert_many_result.inserted_ids)
```

```python
# Find a single document
single_doc = collection.find_one({"name": "Alice"})
print("Single document:", single_doc)
```

```python
# Find multiple documents
multiple_docs = collection.find({"age": {"$gt": 30}})
print("Multiple documents:")
for doc in multiple_docs:
    print(doc)
```

```python
# Update a single document
update_result = collection.update_one({"name": "Alice"}, {"$set": {"age": 26}})
print("Updated document count:", update_result.modified_count)
```

```python
# Update multiple documents
update_many_result = collection.update_many({"age": {"$gt": 30}}, {"$set": {"status": "senior"}})
print("Updated document count:", update_many_result.modified_count)
```

```python
# Delete a single document
delete_result = collection.delete_one({"name": "Alice"})
print("Deleted document count:", delete_result.deleted_count)
```

```python
# Delete multiple documents
delete_many_result = collection.delete_many({"age": {"$lt": 40}})
print("Deleted document count:", delete_many_result.deleted_count)
```
